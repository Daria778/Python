```python
pg = pg.dropna()
```

```python
pg.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data = pg, x = 'bill_length_mm', y = 'bill_depth_mm')
```

```python
pg.loc[(pg['bill_length_mm'] >= 50) & (pg['bill_depth_mm'] >= 18), 'bill_size'] = 'big_bill'
pg.loc[((pg['bill_length_mm'] <= 55) & (pg['bill_length_mm'] >= 45)) & ((pg['bill_depth_mm'] >=17) & (pg['bill_depth_mm'] <= 18)), 'bill_size'] = 'middle_bill'
pg.loc[(pg['bill_length_mm'] <= 45) & (pg['bill_depth_mm'] <= 17), 'bill_size'] = 'small_bill'
```

```python
pg.head()
```

```python
pg['flipper_length_mm'].mean()
```

```python
sns.histplot(data = pg, x = 'bill_size')
```
